- Thành phần dữ liệu: tử số, mẫu số.
- Thành phần xử lý: xuất-nhập giá trị cho phân số, cộng, trừ, nhân, chia hai # phân số, rút gọn, đổi dấu...
```python
class PhanSo:
    def __init__(self, tu, mau):
        self.tu = tu
        self.mau = mau

    def __str__(self):
        return "{}/{}".format(self.tu, self.mau)

    def doi_dau(self):
        return PhanSo(-self.tu, self.mau)

    def rut_gon(self):
        gcd = self._tim_uoc_chung_lon_nhat(self.tu, self.mau)
        return PhanSo(self.tu // gcd, self.mau // gcd)

    def __add__(self, other):
        tu = self.tu * other.mau + self.mau * other.tu
        mau = self.mau * other.mau
        return PhanSo(tu, mau).rut_gon()

    def __sub__(self, other):
        return self + other.doi_dau()

    def __mul__(self, other):
        tu = self.tu * other.tu
        mau = self.mau * other.mau
        return PhanSo(tu, mau).rut_gon()

    def __truediv__(self, other):
        return self * PhanSo(other.mau, other.tu)

    def __eq__(self, other):
        return self.tu * other.mau == self.mau * other.tu

    def __lt__(self, other):
        return self.tu * other.mau < self.mau * other.tu

    def __gt__(self, other):
        return self.tu * other.mau > self.mau * other.tu

    def _tim_uoc_chung_lon_nhat(self, a, b):
        if b == 0:
            return a
        return self._tim_uoc_chung_lon_nhat(b, a % b)
#1. Khởi tạo phân số
a = PhanSo(2, 3)
b = PhanSo(1, 4)
c = PhanSo(3, 1)
d = PhanSo(2, 5)
res = 0
#2. Xuất phân số
print("a = ", a) 
#3. Đổi dấu phân số
print("Doi dau a = ", a.doi_dau())
#4. Rút gọn phân số
print("Rut gon: ", PhanSo(6, 12).rut_gon())
#5. Cộng hai phân số
res = a + b
print("a + b = ", res)
#6. Trừ hai phân số
res = a - b
print("a - b = ", res)
#7. Nhân hai phân số
res = b * c
print("b * c = ", res)
#8. Chia hai phân số
res = c / b
print("c / b", res)
#9. Thực hiện phép so sánh hai phân số.
print(a == b)
print(a > b)
#10. Tìm phân số lớn nhất trong nhiều phân số
ps_list = [a, b, c, d]
max_ps = max(ps_list)
print(max_ps)  
#11. Tự xây dựng thêm các thao tác thích hợp minh họa cho bài toán.
```

Bài 2. Xây dựng lớp ToaDo để biểu diễn một điểm trong mặt phẳng:
- Thành phần dữ liệu: hoành độ và tung độ
- Thành phần xử lý: nhập, xuất, thay đổi tọa độ điểm, lấy tung độ, lấy hoành # đ# ộ, thay đổi tung độ, hoành độ, tịnh tiến...
Viết chương trình minh họa:
1. Khởi tạo tọa độ một điểm
2. Xuất tọa độ một điểm
3. Lấy tung độ, hoành độ của một điểm
4. Thay đổi tung độ, hoành độ của một điểm
5. Di chuyển đến tọa độ
6. Khoảng cách hai điểm
7. Tìm tọa độ trung điểm khi biết hai điểm
8. Xóa tọa độ điểm
9. Sao chép điểm tọa độ

```python
import math

class ToaDo:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0}, {1})".format(self.x, self.y)

    def get_x(self):
        return self.x

    def get_y(self):
        return self.y

    def set_x(self, x):
        self.x = x

    def set_y(self, y):
        self.y = y

    def move_to(self, x, y):
        self.x = x
        self.y = y

    def distance_to(self, other):
        dx = self.x - other.get_x()
        dy = self.y - other.get_y()
        return math.sqrt(dx*dx + dy*dy)

    def midpoint_to(self, other):
        mx = (self.x + other.get_x()) / 2
        my = (self.y + other.get_y()) / 2
        return ToaDo(mx, my)

    def delete(self):
        del self.x
        del self.y

    def clone(self):
        return ToaDo(self.x, self.y)
    
#1. Khởi tạo tọa độ một điểm
p = ToaDo(3, 4)
#2. Xuất tọa độ một điểm
print("A: ", p)
#3. Lấy tung độ, hoành độ của một điểm
print("x = {0}, y = {1}".format(p.get_x(), p.get_y()))
#4. Thay đổi tung độ, hoành độ của một điểm
p.set_x(5)
p.set_y(6)
#5. Di chuyển đến tọa độ
p.move_to(10, 10)
#6. Khoảng cách hai điểm
q = ToaDo(7, 8)
print("Khoang cach: ",p)
dist = p.distance_to(q)
print("Khoang cach cua p va q = {0:.2f}".format(dist))
#7. Tìm tọa độ trung điểm khi biết hai điểm
mid = p.midpoint_to(q)
print("Toa do trung diem p va q = {0}".format(mid))
#8. Xóa tọa độ điểm
p.delete()
#9. Sao chép điểm tọa độ
q_clone = q.clone()
print(q_clone)
```

Bài 3. Thiết kế lớp TamGiac để biểu diễn một tam giác trong mặt phẳng. Sau đó, viết chương trình minh họa:
1. Khởi tạo tam giác
2. Xuất thông tin tam giác
3. Xét loại tam giác
4. Tính chu vi
5. Tính diện tích
6. Tìm bán kính đường tròn ngoại tiếp của tam giác.
7. Tìm bán kính đường tròn nội tiếp của tam giác.

```python
class ToaDo:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def xuat(self):
        print(f"({self.x}, {self.y})")

    def doi_dau(self):
        self.x = -self.x
        self.y = -self.y

    def lay_x(self):
        return self.x

    def lay_y(self):
        return self.y

    def thay_doi_x(self, x):
        self.x = x

    def thay_doi_y(self, y):
        self.y = y

    def tinh_khoang_cach(self, td):
        return ((self.x - td.x) ** 2 + (self.y - td.y) ** 2) ** 0.5


class TamGiac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def xuat(self):
        print(f"Điểm A: ", end="")
        self.a.xuat()
        print(f"Điểm B: ", end="")
        self.b.xuat()
        print(f"Điểm C: ", end="")
        self.c.xuat()

    def loai_tam_giac(self):
        ab = self.a.tinh_khoang_cach(self.b)
        bc = self.b.tinh_khoang_cach(self.c)
        ac = self.a.tinh_khoang_cach(self.c)

        if ab == bc == ac:
            return "Tam giác đều"
        elif ab == bc or ab == ac or bc == ac:
            return "Tam giác cân"
        elif ab**2 == bc**2 + ac**2 or bc**2 == ab**2 + ac**2 or ac**2 == ab**2 + bc**2:
            return "Tam giác vuông"
        else:
            return "Tam giác thường"

    def chu_vi(self):
        ab = self.a.tinh_khoang_cach(self.b)
        bc = self.b.tinh_khoang_cach(self.c)
        ac = self.a.tinh_khoang_cach(self.c)
        return ab + bc + ac

    def dien_tich(self):
        ab = self.a.tinh_khoang_cach(self.b)
        bc = self.b.tinh_khoang_cach(self.c)
        ac = self.a.tinh_khoang_cach(self.c)
        p = (ab + bc + ac) / 2
        return (p * (p - ab) * (p - bc) * (p - ac)) ** 0.5

    def ban_kinh_duong_tron_ngoai_tiep(self):
        ab = self.a.tinh_khoang_cach(self.b)
        bc = self.b.tinh_khoang_cach(self.c)
        ac = self.a.tinh_khoang_cach(self.c)
        p = (ab + bc + ac) / 2
        return (ab * bc * ac) / (4 * (p * (p - ab) * (p - bc) * (p - ac)) ** 0.5)

    def ban_kinh_duong_tron_noi_tiep(self):
        ab = self.a.tinh_khoang_cach(self.b)
        bc = self.b.tinh_khoang_cach(self.c)
        ac = self.a.tinh_khoang_cach(self.c)
        p = self.chu_vi() / 2
        return ((p - ab) * (p - bc) * (p - ac)) / (p * (p - ab - bc - ac) ** 0.5)

#1. Khởi tạo tam giác
tg = TamGiac(ToaDo(0,0), ToaDo(4,0), ToaDo(0,3))
#2. Xuất thông tin tam giác
tg.xuat()
#3. Xét loại tam giác
print("Loai tam giac:", tg.loai_tam_giac())
#4. Tính chu vi
print("Chu vi:", tg.chu_vi())
#5. Tính diện tích
print("Dien tich:", tg.dien_tich())
#6. Tìm bán kính đường tròn ngoại tiếp của tam giác.
print("Ban kinh duong tron ngoai tiep:", tg.ban_kinh_duong_tron_ngoai_tiep())
#7. Tìm bán kính đường tròn nội tiếp của tam giác.
print("Ban kinh duong tron noi tiep:", tg.ban_kinh_duong_tron_noi_tiep())

```

BÀI 4. Xây dựng một lớp để biểu diễn một đường tròn trong mặt phẳng. Sau đó, viết chương trình minh họa:
1. Khởi tạo đường tròn
2. Xuất thông tin đường tròn
3. Tính chu vi
4. Tính diện tích
5. Phóng to đường tròn theo phần trăm
6. Thu nhỏ đường tròn theo phần trăm

```python
import math

class Circle:
    def __init__(self, x, y, r):
        self.x = x
        self.y = y
        self.r = r

    def xuat(self):
        print(f"Đường tròn có tâm ({self.x}, {self.y}) và bán kính {self.r}")

    def chu_vi(self):
        return 2 * math.pi * self.r

    def dien_tich(self):
        return math.pi * self.r ** 2

    def phong_to(self, percent):
        self.r *= (1 + percent / 100)

    def thu_nho(self, percent):
        self.r *= (1 - percent / 100)

# 1. Khởi tạo đường tròn
circle = Circle(0, 0, 5)
# 2. Xuất thông tin đường tròn
circle.xuat()
# 3. Tính chu vi
print("Chu vi:", circle.chu_vi())
# 4. Tính diện tích
print("Diện tích:", circle.dien_tich())
# 5. Phóng to đường tròn theo phần trăm
circle.phong_to(50)
circle.xuat()
# 6. Thu nhỏ đường tròn theo phần trăm
circle.thu_nho(25)
circle.xuat()
```

BÀI 5. Xây dựng lớp nhân viên (NhanVien) cần quản lý các thông tin sau: mã nhân viên, họ tên, lương cơ bản, số sản phẩm đã bán được, lương hằng tháng. Lương hằng tháng cho nhân viên được tính theo công thức sau:
- Lương HT = lương căn bản + số sản phẩm * 175.000 đ
- Nếu lương từ 10 triệu trở lên thì thưởng thêm 10%.
Sau đó, viết chương trình theo giải quyết các yêu cầu sau:
1. Tạo dữ liệu nhân viên.
2. Xuất các nhân viên trong công ty.
3. Tính lương nhân viên.
4. Tìm nhân viên theo theo mã nhân viên.
5. Cập nhật lương cơ bản nhân viên theo mã nhân viên.
6. Tìm nhân viên có lương cao nhất.
7. Tìm nhân viên có số sản phẩm bán được thấp nhất.
8. Tìm 10 nhân viên có lương cao nhất.
9. Tự xây dựng thêm các thao tác thích hợp cho bài toán.

```python
class NhanVien:
    def __init__(self, ma_nv, ho_ten, luong_cb, so_sp_ban):
        self.ma_nv = ma_nv
        self.ho_ten = ho_ten
        self.luong_cb = luong_cb
        self.so_sp_ban = so_sp_ban
        self.luong_ht = self.tinh_luong_ht()

    def tinh_luong_ht(self):
        luong_ht = self.luong_cb + self.so_sp_ban * 175000
        if luong_ht >= 10000000:
            luong_ht += luong_ht * 0.1
        return luong_ht

    def xuat(self):
        print(f"Mã nhân viên: {self.ma_nv}")
        print(f"Họ và tên: {self.ho_ten}")
        print(f"Lương cơ bản: {self.luong_cb}")
        print(f"Số sản phẩm bán được: {self.so_sp_ban}")
        print(f"Lương hằng tháng: {self.luong_ht}")

    def cap_nhat_luong_cb(self, luong_cb_moi):
        self.luong_cb = luong_cb_moi
        self.luong_ht = self.tinh_luong_ht()
    
    def __lt__(self, other):
        return self.luong_ht < other.luong_ht
    
    
# 1. Tạo dữ liệu nhân viên
nv1 = NhanVien("NV001", "Nguyễn Văn A", 8000000, 20)
nv2 = NhanVien("NV002", "Trần Thị B", 10000000, 30)
nv3 = NhanVien("NV003", "Phạm Văn C", 6000000, 15)
nv4 = NhanVien("NV004", "Hoàng Thị D", 12000000, 50)
nv5 = NhanVien("NV005", "Lê Văn E", 9000000, 25)
ds_nv = [nv1, nv2, nv3, nv4, nv5]

# 2. Xuất các nhân viên trong công ty
print("Danh sách nhân viên:")
for nv in ds_nv:
    nv.xuat()

# 3. Tính lương nhân viên
print("Lương nhân viên:")
for nv in ds_nv:
    print(f"{nv.ho_ten}: {nv.luong_ht}")

# 4. Tìm nhân viên theo theo mã nhân viên
ma_nv = "NV002"
print(f"Tìm nhân viên có mã {ma_nv}:")
for nv in ds_nv:
    if nv.ma_nv == ma_nv:
        nv.xuat()

# 5. Cập nhật lương cơ bản nhân viên theo mã nhân viên
ma_nv = "NV003"
luong_cb_moi = 6500000
print(f"Cập nhật lương cơ bản của nhân viên có mã {ma_nv} thành {luong_cb_moi}:")

# 6. Tìm nhân viên có lương cao nhất
max_luong_nv = max(ds_nv)
print(f"Nhân viên có lương cao nhất: {max_luong_nv.ho_ten}, Lương: {max_luong_nv.luong_ht}")

# 7. Tìm nhân viên có số sản phẩm bán được thấp nhất
min_so_sp_nv = min(ds_nv, key=lambda nv: nv.so_sp_ban)
print(f"Nhân viên có số sản phẩm bán được thấp nhất: {min_so_sp_nv.ho_ten}, Số sản phẩm bán được: {min_so_sp_nv.so_sp_ban}")

# 8. Tìm 10 nhân viên có lương cao nhất
top10_luong_nv = sorted(ds_nv, reverse=True)[:10]
print("Top 10 nhân viên có lương cao nhất:")
for i, nv in enumerate(top10_luong_nv):
    print(f"{i+1}. {nv.ho_ten}, Lương: {nv.luong_ht}")

# 9. Sắp xếp danh sách nhân viên theo lương giảm dần và in ra màn hình
ds_nv_theo_luong = sorted(ds_nv, reverse=True, key=lambda nv: nv.luong_ht)
print("Danh sách nhân viên theo lương giảm dần:")
for nv in ds_nv_theo_luong:
    print(f"{nv.ho_ten}, Lương: {nv.luong_ht}")
```
