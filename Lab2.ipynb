- Thành phần dữ liệu: tử số, mẫu số.
- Thành phần xử lý: xuất-nhập giá trị cho phân số, cộng, trừ, nhân, chia hai # phân số, rút gọn, đổi dấu... Sau đó, thiết kế chương trình minh họa các yêu cầu sau:

2. Xuất phân số
3. Đổi dấu phân số
4. Rút gọn phân số

```python
class PhanSo:
    def __init__(self, tu, mau):
        self.tu = tu
        self.mau = mau

    def __str__(self):
        return "{}/{}".format(self.tu, self.mau)

    def doi_dau(self):
        return PhanSo(-self.tu, self.mau)

    def rut_gon(self):
        gcd = self._tim_uoc_chung_lon_nhat(self.tu, self.mau)
        return PhanSo(self.tu // gcd, self.mau // gcd)

    def __add__(self, other):
        tu = self.tu * other.mau + self.mau * other.tu
        mau = self.mau * other.mau
        return PhanSo(tu, mau).rut_gon()

    def __sub__(self, other):
        return self + other.doi_dau()

    def __mul__(self, other):
        tu = self.tu * other.tu
        mau = self.mau * other.mau
        return PhanSo(tu, mau).rut_gon()

    def __truediv__(self, other):
        return self * PhanSo(other.mau, other.tu)

    def __eq__(self, other):
        return self.tu * other.mau == self.mau * other.tu

    def __lt__(self, other):
        return self.tu * other.mau < self.mau * other.tu

    def __gt__(self, other):
        return self.tu * other.mau > self.mau * other.tu

    def _tim_uoc_chung_lon_nhat(self, a, b):
        if b == 0:
            return a
        return self._tim_uoc_chung_lon_nhat(b, a % b)
#1. Khởi tạo phân số
a = PhanSo(2, 3)
b = PhanSo(1, 4)
c = PhanSo(3, 1)
d = PhanSo(2, 5)
res = 0
#2. Xuất phân số
print("a = ", a) 
#3. Đổi dấu phân số
print("Doi dau a = ", a.doi_dau())
#4. Rút gọn phân số
print("Rut gon: ", PhanSo(6, 12).rut_gon())
#5. Cộng hai phân số
res = a + b
print("a + b = ", res)
#6. Trừ hai phân số
res = a - b
print("a - b = ", res)
#7. Nhân hai phân số
res = b * c
print("b * c = ", res)
#8. Chia hai phân số
res = c / b
print("c / b", res)
#9. Thực hiện phép so sánh hai phân số.
print(a == b)
print(a > b)
#10. Tìm phân số lớn nhất trong nhiều phân số
ps_list = [a, b, c, d]
max_ps = max(ps_list)
print(max_ps)  
#11. Tự xây dựng thêm các thao tác thích hợp minh họa cho bài toán.
```

Bài 2. Xây dựng lớp ToaDo để biểu diễn một điểm trong mặt phẳng:
- Thành phần dữ liệu: hoành độ và tung độ
- Thành phần xử lý: nhập, xuất, thay đổi tọa độ điểm, lấy tung độ, lấy hoành # đ# ộ, thay đổi tung độ, hoành độ, tịnh tiến...
Viết chương trình minh họa:
1. Khởi tạo tọa độ một điểm
2. Xuất tọa độ một điểm
3. Lấy tung độ, hoành độ của một điểm
4. Thay đổi tung độ, hoành độ của một điểm
5. Di chuyển đến tọa độ
6. Khoảng cách hai điểm
7. Tìm tọa độ trung điểm khi biết hai điểm
8. Xóa tọa độ điểm
9. Sao chép điểm tọa độ

```python
import math

class ToaDo:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0}, {1})".format(self.x, self.y)

    def get_x(self):
        return self.x

    def get_y(self):
        return self.y

    def set_x(self, x):
        self.x = x

    def set_y(self, y):
        self.y = y

    def move_to(self, x, y):
        self.x = x
        self.y = y

    def distance_to(self, other):
        dx = self.x - other.get_x()
        dy = self.y - other.get_y()
        return math.sqrt(dx*dx + dy*dy)

    def midpoint_to(self, other):
        mx = (self.x + other.get_x()) / 2
        my = (self.y + other.get_y()) / 2
        return ToaDo(mx, my)

    def delete(self):
        del self.x
        del self.y

    def clone(self):
        return ToaDo(self.x, self.y)
    
#1. Khởi tạo tọa độ một điểm
#2. Xuất tọa độ một điểm
#3. Lấy tung độ, hoành độ của một điểm
#4. Thay đổi tung độ, hoành độ của một điểm
#5. Di chuyển đến tọa độ
#6. Khoảng cách hai điểm
#7. Tìm tọa độ trung điểm khi biết hai điểm
#8. Xóa tọa độ điểm
#9. Sao chép điểm tọa độ
```
