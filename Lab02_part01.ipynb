# Bài 1. Xây dựng lớp PhanSo để biểu diễn một phân số gồm hai thành phần:
# - Thành phần dữ liệu: tử số, mẫu số.
# - Thành phần xử lý: xuất-nhập giá trị cho phân số, cộng, trừ, nhân, chia hai # phân số, rút gọn, đổi dấu... Xây dựng các phương thức xử lý, nạp chồng các phép # toán +, -, *, /, ==, !=, >;, >=, <, <=#  trên phân số, nạp chồng hàm print. Ví # dụ: phân số + phân số, hoặc phân số + một số... Code ví dụ:
# a = PhanSo(2,3), b = PhanSo(1,4), c = PhanSo(3,1), d = PhanSo (2,5)
# a = b * -c
# c = (b+2) * 2/a
# d = a/3 + (b*c-2)/5
# print(c), print(d)
# Sau đó, thiết kế chương trình minh họa các yêu cầu sau:
# 1. Khởi tạo phân số
# 2. Xuất phân số
# 3. Đổi dấu phân số
# 4. Rút gọn phân số
# 5. Cộng hai phân số
# 6. Trừ hai phân số
# 7. Nhân hai phân số
# 8. Chia hai phân số
# 9. Thực hiện phép so sánh hai phân số.
# 10. Tìm phân số lớn nhất trong nhiều phân số
# 11. Tự xây dựng thêm các thao tác thích hợp minh họa cho bài toán.

```python
class PhanSo:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau

    def __str__(self):
        return str(self.tu) + '/' + str(self.mau)

    def doi_dau(self):
        self.tu = -self.tu

    def rut_gon(self):
        ucln = self.UCLN(self.tu, self.mau)
        self.tu //= ucln
        self.mau //= ucln

    def __add__(self, other):
        tu_moi = self.tu*other.mau + other.tu*self.mau
        mau_moi = self.mau * other.mau
        return PhanSo(tu_moi, mau_moi)

    def __sub__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        mau_moi = self.mau * other.mau
        return PhanSo(tu_moi, mau_moi)

    def __mul__(self, other):
        tu_moi = self.tu * other.tu
        mau_moi = self.mau * other.mau
        return PhanSo(tu_moi, mau_moi)

    def __truediv__(self, other):
        tu_moi = self.tu * other.mau
        mau_moi = self.mau * other.tu
        return PhanSo(tu_moi, mau_moi)

    def __eq__(self, other):
        self.rut_gon()
        other.rut_gon()
        return self.tu == other.tu and self.mau == other.mau

    def __ne__(self, other):
        return not self == other

    def __gt__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        return tu_moi > 0

    def __ge__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        return tu_moi >= 0

    def __lt__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        return tu_moi < 0

    def __le__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        return tu_moi <= 0

    def UCLN(self, a, b):
        if b == 0:
            return a
        return self.UCLN(b, a%b)
    
    def to_decimal(self):
        return self.tu / self.mau
    
    def inverse(self):
        return PhanSo(self.mau, self.tu)
    
    def power(self, n):
        return PhanSo(self.tu ** n, self.mau ** n)
    
"""1. Khởi tạo phân số"""
a = PhanSo(2,3)
b = PhanSo(1,4)
"""2. Xuất phân số"""
print("a = ", a)
print("b = ", b)
"""3. Đổi dấu phân số"""

"""4. Rút gọn phân số"""
c = PhanSo(12, 24)
c.rut_gon()
print("c = ", c)
"""5. Cộng hai phân số"""
d = a + b
print("a + b = ", d)
"""6. Trừ hai phân số"""
e=a-b
print("a - b = ",e)
"""7. Nhân hai phân số"""
f=a*b
print("a * b = ",f)
"""8. Chia hai phân số"""
g=a/b
print("a / b = ",g)
"""9. Thực hiện phép so sánh hai phân số."""
print("a > b = ",a>b)
"""10. Tìm phân số lớn nhất trong nhiều phân số"""
ps1 = PhanSo(2, 3)
ps2 = PhanSo(3, 4)
ps3 = PhanSo(1, 2)

max_ps = ps1

if ps2 > max_ps:
    max_ps = ps2

if ps3 > max_ps:
    max_ps = ps3

print("Phân số lớn nhất là:", max_ps)

"""11. Tự xây dựng thêm các thao tác thích hợp minh họa cho bài toán."""
ps = PhanSo(2, 3)
print("Phân số:", ps)
print("Số thập phân:", ps.to_decimal())
print("Nghịch đảo của phân số:", ps.inverse())
print("Lũy thừa bậc 3 của phân số:", ps.power(3))
```

# Bài 2. Xây dựng lớp ToaDo để biểu diễn một điểm trong mặt phẳng:
# - Thành phần dữ liệu: hoành độ và tung độ
# - Thành phần xử lý: nhập, xuất, thay đổi tọa độ điểm, lấy tung độ, lấy hoành # đ# ộ, thay đổi tung độ, hoành độ, tịnh tiến...
# Viết chương trình minh họa:
# 1. Khởi tạo tọa độ một điểm
# 2. Xuất tọa độ một điểm
# 3. Lấy tung độ, hoành độ của một điểm
# 4. Thay đổi tung độ, hoành độ của một điểm
# 5. Di chuyển đến tọa độ
# 6. Khoảng cách hai điểm
# 7. Tìm tọa độ trung điểm khi biết hai điểm
# 8. Xóa tọa độ điểm
# 9. Sao chép điểm tọa độ
# 10. Tìm tọa độ có khoảng cách gần tâm O nhất trong nhiều tọa độ. Nếu có nhiều tọa độ thì lấy tọa độ đầu tiên được tìm thấy.

```python
import math

class ToaDo:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def nhap(self):
        self.x = float(input("Nhap hoanh do: "))
        self.y = float(input("Nhap tung do: "))

    def xuat(self):
        print(f"({self.x}, {self.y})")

    def lay_tung_do(self):
        return self.y

    def lay_hoanh_do(self):
        return self.x

    def thay_doi_tung_do(self, dy):
        self.y += dy

    def thay_doi_hoanh_do(self, dx):
        self.x += dx

    def dich_chuyen_toa_do(self, dx, dy):
        self.x += dx
        self.y += dy

    def khoang_cach_hai_diem(self, p):
        return math.sqrt((self.x - p.x) ** 2 + (self.y - p.y) ** 2)

    def toa_do_trung_diem(self, p):
        return ToaDo((self.x + p.x) / 2, (self.y + p.y) / 2)

    def xoa(self):
        self.x = 0
        self.y = 0

    def sao_chep(self):
        return ToaDo(self.x, self.y)

    def tim_toa_do_gan_O_nhat(self, toado_list):
        min_distance = float('inf')
        closest_toado = None
        for toado in toado_list:
            distance = self.khoang_cach_hai_diem(toado)
            if distance < min_distance:
                min_distance = distance
                closest_toado = toado
        return closest_toado
    
"""1. Khởi tạo tọa độ một điểm"""
p1 = ToaDo(3, 4)
"""2. Xuất tọa độ một điểm"""
p1.xuat() 
"""3. Lấy tung độ, hoành độ của một điểm"""
y = p1.lay_tung_do()
x = p1.lay_hoanh_do()
print("p1 có x = ",x,", y = ",y)
"""4. Thay đổi tung độ, hoành độ của một điểm"""
p1.thay_doi_tung_do(2)
p1.thay_doi_hoanh_do(-1)
print("p1 sau khi thay đổi tung độ và hoành độ: ")
p1.xuat()
"""5. Di chuyển đến tọa độ"""
p1.dich_chuyen_toa_do(2, -3)
print("Toa độ p1 đã di chuyển là: ")
p1.xuat()
"""6. Khoảng cách hai điểm"""
p2 = ToaDo(5, 2)
kc = p1.khoang_cach_hai_diem(p2)
print("Khoảng cách p1 so với p2 = ",kc)
"""7. Tìm tọa độ trung điểm khi biết hai điểm"""
p3 = ToaDo(8, 6)
td = p1.toa_do_trung_diem(p3)
print("Toạ độ trung điểm: ")
td.xuat()
"""8. Xóa tọa độ điểm"""
p1.xoa()
print("Toạ độ p1 đã được xoá")
"""9. Sao chép điểm tọa độ"""
p4 = p2.sao_chep()
p4.xuat()
"""10. Tìm tọa độ có khoảng cách gần tâm O nhất trong nhiều tọa độ. 
Nếu có nhiều tọa độ thì lấy tọa độ đầu tiên được tìm thấy."""
toado_list = [ToaDo(1,2), ToaDo(3,4), ToaDo(-1,2)]
O = ToaDo(0,0)
closest_toado = O.tim_toa_do_gan_O_nhat(toado_list)
print("Tọa độ gần tâm O nhất là: ({}, {})".format(closest_toado.x, closest_toado.y))
```

# Bài 3. Thiết kế lớp TamGiac để biểu diễn một tam giác trong mặt phẳng. Sau đó, viết chương trình minh họa:
# 1. Khởi tạo tam giác
# 2. Xuất thông tin tam giác
# 3. Xét loại tam giác
# 4. Tính chu vi
# 5. Tính diện tích
# 6. Tìm bán kính đường tròn ngoại tiếp của tam giác.
# 7. Tìm bán kính đường tròn nội tiếp của tam giác.

```python
class TamGiac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def xuat_thong_tin(self):
        print("Độ dài các cạnh tam giác: a = {0}, b = {1}, c = {2}"
              .format(self.a, self.b, self.c))
        
    def loai_tam_giac(self):
        if self.a == self.b == self.c:
            return "Tam giác đều"
        elif self.a == self.b or self.b == self.c or self.c == self.a:
            if self.a**2 + self.b**2 == self.c**2 or self.a**2 + self.c**2 == self.b**2 or self.b**2 + self.c**2 == self.a**2:
                return "Tam giác vuông cân"
            else:
                return "Tam giác cân"
        elif self.a**2 + self.b**2 == self.c**2 or self.a**2 + self.c**2 == self.b**2 or self.b**2 + self.c**2 == self.a**2:
            return "Tam giác vuông"
        else:
            return "Tam giác thường"

    def chu_vi(self):
        return self.a + self.b + self.c

    def dien_tich(self):
        p = self.chu_vi() / 2
        return (p * (p - self.a) * (p - self.b) * (p - self.c)) ** 0.5

    def ban_kinh_ngoai_tiep(self):
        return (self.a * self.b * self.c) / (4 * self.dien_tich())

    def ban_kinh_noi_tiep(self):
        return (self.dien_tich() * 2) / (self.a + self.b + self.c)
    
    def __str__(self):
        return f"Tam giác có ba cạnh là {self.a}, {self.b}, {self.c}."

"""1. Khởi tạo tam giác"""
tg = TamGiac(3, 4, 5)
"""2. Xuất thông tin tam giác"""
tg.xuat_thong_tin()
"""3. Xét loại tam giác"""
print("Tam giác này thuộc ", tg.loai_tam_giac())
"""4. Tính chu vi"""
print("Chu vi:", tg.chu_vi())
"""5. Tính diện tích"""
print("Diện tích:", tg.dien_tich())
"""6. Tìm bán kính đường tròn ngoại tiếp của tam giác."""
print("Bán kính đường tròn ngoại tiếp:", tg.ban_kinh_ngoai_tiep())
"""7. Tìm bán kính đường tròn nội tiếp của tam giác."""
print("Bán kính đường tròn nội tiếp:", tg.ban_kinh_noi_tiep())
```

# BÀI 4. Xây dựng một lớp để biểu diễn một đường tròn trong mặt phẳng. Sau đó, viết chương trình minh họa:
# 1. Khởi tạo đường tròn
# 2. Xuất thông tin đường tròn
# 3. Tính chu vi
# 4. Tính diện tích
# 5. Phóng to đường tròn theo phần trăm
# 6. Thu nhỏ đường tròn theo phần trăm

```python
import math

class Circle:
    def __init__(self, x, y, r):
        self.x = x
        self.y = y
        self.r = r
        
    def get_area(self):
        return math.pi * self.r ** 2
    
    def get_perimeter(self):
        return 2 * math.pi * self.r
    
    def zoom(self, percent):
        self.r *= percent / 100.0
        
    def shrink(self, percent):
        self.r *= 100.0 / percent
        
    def __str__(self):
        return "Circle at ({}, {}) with radius {}".format(self.x, self.y, self.r)

"""1. Khởi tạo đường tròn"""
c = Circle(0, 0, 5)
"""2. Xuất thông tin đường tròn"""
print(c)
"""3. Tính chu vi"""
print("Chu vi hình tròn:", c.get_perimeter())
"""4. Tính diện tích"""
print("Diện tích hình tròn:", c.get_area()) 
"""5. Phóng to đường tròn theo phần trăm"""
c.shrink(200)
print(c) 
"""6. Thu nhỏ đường tròn theo phần trăm"""
c.zoom(50)
print(c) 
```

# BÀI 5. Xây dựng lớp nhân viên (NhanVien) cần quản lý các thông tin sau: mã nhân viên, họ tên, lương cơ bản, số sản phẩm đã bán được, lương hằng tháng. Lương hằng tháng cho nhân viên được tính theo công thức sau:
# - Lương HT = lương căn bản + số sản phẩm * 175.000 đ
# - Nếu lương từ 10 triệu trở lên thì thưởng thêm 10%.
# Sau đó, viết chương trình theo giải quyết các yêu cầu sau:
# 1. Tạo dữ liệu nhân viên.
# 2. Xuất các nhân viên trong công ty.
# 3. Tính lương nhân viên.
# 4. Tìm nhân viên theo theo mã nhân viên.
# 5. Cập nhật lương cơ bản nhân viên theo mã nhân viên.
# 6. Tìm nhân viên có lương cao nhất.
# 7. Tìm nhân viên có số sản phẩm bán được thấp nhất.
# 8. Tìm 10 nhân viên có lương cao nhất.
# 9. Tự xây dựng thêm các thao tác thích hợp cho bài toán.

```python
class NhanVien:
    def __init__(self, maNV, hoTen, luongCB, soSP):
        self.maNV = maNV
        self.hoTen = hoTen
        self.luongCB = luongCB
        self.soSP = soSP
        self.luongHT = self.tinh_luong()

    def tinh_luong(self):
        luongHT = self.luongCB + self.soSP * 175000
        if luongHT >= 10000000:
            luongHT *= 1.1
        return luongHT

    def cap_nhat_luongCB(self, luongCB_moi):
        self.luongCB = luongCB_moi
        self.luongHT = self.tinh_luong()

    def xuat(self):
        print(f"Mã nhân viên: {self.maNV}")
        print(f"Họ tên: {self.hoTen}")
        print(f"Lương cơ bản: {self.luongCB}")
        print(f"Số sản phẩm đã bán: {self.soSP}")
        print(f"Lương hàng tháng: {self.luongHT}")

        
def timNV(ls,ma):
    flag = False
    for i in ls:
        if i.maNV == ma:
            flag = True
            i.xuat()
            print()
    if flag == False:
        print("Không tìm thấy nhân viên.")
        
"""1. Tạo dữ liệu nhân viên."""
nv1 = NhanVien("NV001", "Tran Van A", 8000000, 50)
nv2 = NhanVien("NV002", "Nguyen Thi B", 10000000, 100)
nv3 = NhanVien("NV003", "Le Van C", 5000000, 20)
lsNV = (nv1,nv2,nv3)
"""2. Xuất các nhân viên trong công ty."""
for i in lsNV:
    i.xuat()
    print()
"""3. Tính lương nhân viên."""
print(f"Lương của {nv1.hoTen} la {nv1.luongHT}")
print(f"Lương của {nv2.hoTen} la {nv2.luongHT}")
print(f"Lương của {nv3.hoTen} la {nv3.luongHT}")
print()
"""4. Tìm nhân viên theo theo mã nhân viên."""
mnv = input("Nhập mã nhập viên cần tìm:")
timNV(lsNV,mnv)
print()
"""5. Cập nhật lương cơ bản nhân viên theo mã nhân viên."""
nv1.cap_nhat_luongCB(9000000)
nv1.xuat()
print()
"""6. Tìm nhân viên có lương cao nhất."""
luong_cao_nhat = max(lsNV, key=lambda x: x.luongHT)
print(f"Nhân viên có lương cao nhất là {luong_cao_nhat.hoTen} với lương là {luong_cao_nhat.luongHT}")
print()
"""7. Tìm nhân viên có số sản phẩm bán được thấp nhất."""
sp_thap_nhat = min(lsNV, key=lambda x: x.soSP)
print(f"Nhân viên có số sản phẩm bán ra thấp nhất là {sp_thap_nhat.hoTen} với số sản phẩm là {sp_thap_nhat.soSP} ")
print()
"""8. Tìm 10 nhân viên có lương cao nhất."""
sorted_nhanvien = sorted(lsNV, key=lambda x: x.tinh_luong(), reverse=True)
top10_nhanvien = sorted_nhanvien[:10]
print("10 nhân viên có mức lương cao nhất là: ")
for nhanvien in top10_nhanvien:
    print(nhanvien.maNV, nhanvien.hoTen, nhanvien.tinh_luong())
    print()
"""9. Tự xây dựng thêm các thao tác thích hợp cho bài toán."""
def tinh_tong_luong_cong_ty(danh_sach_nhan_vien):
    tong_luong = 0
    for nhan_vien in danh_sach_nhan_vien:
        tong_luong += nhan_vien.tinh_luong()
    return tong_luong

def tinh_trung_binh_luong(danh_sach_nhan_vien):
    tong_luong = tinh_tong_luong_cong_ty(danh_sach_nhan_vien)
    so_nhan_vien = len(danh_sach_nhan_vien)
    if so_nhan_vien > 0:
        return tong_luong / so_nhan_vien
    else:
        return 0

#Tổng lương công ty:
print("Tổng lương của công ty: " + str(tinh_tong_luong_cong_ty(lsNV)))
print()
#Lương trung bình của công ty:
print("Lương trung bình của công ty: " + str(tinh_trung_binh_luong(lsNV)))
```
