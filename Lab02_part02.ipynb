# Bài 1. Xây dựng chương trình tính tiền lương cho các nhân viên trong 1 công ty. Biết công ty có 2 loại nhân viên: Nhân viên văn phòng, nhân viên kinh doanh. Mỗi nhân viên cần quản lý các thông tin sau: mã nhân viên, họ tên, lương cơ bản, lương hằng tháng.
# Công thức tính tiền hằng tháng cho các nhân viên như sau:
# * Đối với nhân viên kinh doanh:
# Lương = Lương căn bản + số sản phẩm bán ra * 120.000đ
# * Đối với nhân viên văn phòng:
# Lương = Lương căn bản + số ngày làm việc 8 180.000đ
# Nếu lương hằng tháng hơn 8.000.000 thì được thưởng thêm 5%, nếu nhân viên kinh doanh có lương hằng tháng dưới 5.000.000 thì phụ cấp thêm 30%
# Áp dụng OOP và tính trừu tượng định nghĩa các lớp( Vận dụng kế thừa, đa hình, trừu tượng). Đáp ứng các nhu cầu sau:

```python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, maNV, hoTen, luongCoBan):
        self.maNV = maNV
        self.hoTen = hoTen
        self.luongCoBan = luongCoBan
    
    @abstractmethod
    def tinh_luong_hang_thang(self):
        pass
    @abstractmethod
    def in_thong_tin(self):
        pass
    @abstractmethod
    def in_thong_tin_voi_luong(self):
        pass
    
class SalesEmployee(Employee):
    def __init__(self, maNV, hoTen, luongCoBan, soSanPhamBanRa):
        super().__init__(maNV, hoTen, luongCoBan)
        self.soSanPhamBanRa = soSanPhamBanRa

    def tinh_luong_hang_thang(self):
        luong = self.luongCoBan + self.soSanPhamBanRa * 120000
        if luong < 5000000:
            luong *= 1.3
        elif luong > 8000000:
            luong *= 1.05
        return luong
    
    def in_thong_tin(self):
        print("Mã nhân viên:", self.maNV)
        print("Họ tên:", self.hoTen)
        print("Lương cơ bản:", self.luongCoBan)
        print("Số sản phẩm bán ra:", self.soSanPhamBanRa)

    def in_thong_tin_voi_luong(self):
        print("Mã nhân viên:", self.maNV)
        print("Họ tên:", self.hoTen)
        print("Lương cơ bản:", self.luongCoBan)
        print("Số sản phẩm bán ra:", self.soSanPhamBanRa)
        print("Lương hàng tháng NV Sales:", self.tinh_luong_hang_thang())
        
class OfficeEmployee(Employee):
    def __init__(self, maNV, hoTen, luongCoBan, soNgayLamViec):
        super().__init__(maNV, hoTen, luongCoBan)
        self.soNgayLamViec = soNgayLamViec

    def tinh_luong_hang_thang(self):
        luong = self.luongCoBan + self.soNgayLamViec * 180000
        if luong < 5000000:
            luong *= 1.3
        elif luong > 8000000:
            luong *= 1.05
        return luong
    
    def in_thong_tin(self):
        print("Mã nhân viên:", self.maNV)
        print("Họ tên:", self.hoTen)
        print("Lương cơ bản:", self.luongCoBan)
        print("Số giờ làm việc:", self.soNgayLamViec)

    def in_thong_tin_voi_luong(self):
        print("Mã nhân viên:", self.maNV)
        print("Họ tên:", self.hoTen)
        print("Lương cơ bản:", self.luongCoBan)
        print("Số giờ làm việc:", self.soNgayLamViec)
        print("Lương hàng tháng NV Office:", self.tinh_luong_hang_thang())
        
"""1. Xuất danh sách các nhân viên trong công ty"""
nv_kinh_doanh1 = SalesEmployee("NV001", "Nguyen Van A", 5000000, 600)
nv_kinh_doanh2 = SalesEmployee("NV002", "Banh Thi B", 7000000, 2000)
nv_kinh_doanh3 = SalesEmployee("NV003", "Nguyen Thi C", 10000000, 3000)
nv_kinh_doanh4 = SalesEmployee("NV004", "Phan Van D", 4000000, 400)
nv_kinh_doanh5 = SalesEmployee("NV005", "Bach Thi E", 8000000, 800)

nv_van_phong6 = OfficeEmployee("NV006", "Tran Van F", 5000000, 100)
nv_van_phong7 = OfficeEmployee("NV007", "Lâm Ngọc G", 7000000, 200)
nv_van_phong8 = OfficeEmployee("NV008", "Trần Quốc H", 10000000, 300)
nv_van_phong9 = OfficeEmployee("NV009", "Trần Hưng J", 4000000, 150)
nv_van_phong10 = OfficeEmployee("NV010", "Nguyễn K", 8000000, 175)

dsNV = [nv_kinh_doanh1,
        nv_kinh_doanh2,
        nv_kinh_doanh3,
        nv_kinh_doanh4,
        nv_kinh_doanh5,
        nv_van_phong6,
        nv_van_phong7,
        nv_van_phong8,
        nv_van_phong9,
        nv_van_phong10]

for i in dsNV:
    i.in_thong_tin()
    print()

"""2. Tính và cập nhật tiền lương cho các loại nhân viên trong công ty"""
for i in dsNV:
    i.in_thong_tin_voi_luong()
    print()
    
"""3. Tìm nhân viên theo mã nhân viên"""

iD = input("Nhập mã nhân viên cần tìm: ")
flag = False
for i in dsNV:
    if(i.maNV == iD):
        flag = True
        i.in_thong_tin()
        print()
if(flag == False):
    print("Không tìm thấy nhân viên.")
    
"""4. Tìm các nhân viên có lương thấp nhất"""
def findMinimumSalary(dsNV):
    minSalary = dsNV[0].tinh_luong_hang_thang()
    for i in dsNV:
        if(i.tinh_luong_hang_thang()<minSalary):
            minSalary = i.tinh_luong_hang_thang()
    return minSalary

print ("Danh sách các nhân viên có lương thấp nhất:")
minRS = findMinimumSalary(dsNV)
for i in dsNV:
    if(i.tinh_luong_hang_thang() == minRS):
        i.in_thong_tin_voi_luong()

"""5.Tìm các nhân viên có lương cao nhất"""
def findMaximumSalary(dsNV):
    maxSalary = 0
    for i in dsNV:
        if(i.tinh_luong_hang_thang()>maxSalary):
            maxSalary = i.tinh_luong_hang_thang()
    return maxSalary

print ("Danh sách các nhân viên có lương cao nhất:")
maxRS = findMaximumSalary(dsNV)
for i in dsNV:
    if(i.tinh_luong_hang_thang() == maxRS):
        i.in_thong_tin_voi_luong()
        print()
        
"""6. Top 10 nhân viên có lương cao nhất"""
dsNVSorted = sorted(dsNV, key=lambda x: x.tinh_luong_hang_thang(), reverse=True)
top_10_NV = dsNVSorted[:10]
print("Top 10 nhân viên có lương cao nhất:")
for i in top_10_NV:
    i.in_thong_tin_voi_luong()
    print()
    
"""7. Tự xây dựng thêm các thao tác thích hợp cho bài toán"""
"""Nhân viên có số giờ làm việc nhiều nhất:"""
def findMaxWorkHours(dsNV):
    maxH = 0
    for i in dsNV:
        if type(i) == OfficeEmployee:
            if(i.soNgayLamViec > maxH):
                maxH = i.soNgayLamViec
    return maxH

maxHours = findMaxWorkHours(dsNV)
for i in dsNV:
    if type(i) == OfficeEmployee:
        if(i.soNgayLamViec == maxHours):
            i.in_thong_tin_voi_luong()
            print()

"""Nhân viên có số sản phẩm bán ra nhiều nhất:"""
def findMaxSales(dsNV):
    maxS = 0
    for i in dsNV:
        if type(i) == SalesEmployee:
            if(i.soSanPhamBanRa > maxS):
                maxS = i.soSanPhamBanRa
    return maxS

maxSales = findMaxSales(dsNV)
for i in dsNV:
    if type(i) == SalesEmployee:
        if(i.soSanPhamBanRa == maxSales):
            i.in_thong_tin_voi_luong()
            print()
```

# Bài 2. Viết chương trình tính tiền lương và quản lý nhân viên cho một trung tâm công nghệ. Trung tâm cần quản lý và tính lương cho 3 loại đối tượng nhân viên: Chuyên viên, Nghiên cứu viên, Quản lý. Mỗi đối tượng có các thông tin sau:
# • Chuyên viên: mã nhân viên, họ tên, lương cơ bản, hệ số trách nhiệm, số giờ tăng ca.
# • Nghiên cứu viên: mã nhân viên, họ tên, lương cơ bản, hệ số trách nhiệm, số lượng công trình phát minh.
# • Quản lý: mã nhân viên, họ tên, lương cơ bản, hệ số trách nhiệm, hệ số kiêm nhiệm.
# Tiền lương được tính theo công thức sau:
# • Lương chuyên viên: Lương cơ bản + Lương cơ bản * hệ số trách nhiệm + số giờ tăng ca * 180.000đ.
# • Lương nghiên cứu viên: Lương cơ bản + Lương cơ bản * (hệ số trách nhiệm – 0.2) + số lượng công trình phát minh * 5.500.000đ.
# • Lương quản lý: Lương cơ bản * 0.7 + Lương cơ bản * hệ số trách nhiệm + Lương cơ bản * hệ số kiêm nghiệm.
# Áp dụng OOP trong Python định nghĩa các lớp (vận dụng kế thừa, đa hình, trừu tượng), xây dựng chương trình minh họa các yêu cầu sau:
# Câu 1. Xuất thông tin của các nhân viên (mã nhân viên, họ tên, lương cơ bản, hệ số trách nhiệm, tổng tiền lương), thông tin mỗi nhân viên xuất trên 1 dòng.
# Câu 2. Thực hiện việc tính lương cho từng nhân viên.
# Câu 3. Tìm kiếm nhân viên theo mã nhân viên.
# Câu 4. Tính tổng số tiền phải trả cho tất cả nhân viên.
# Câu 5. Tìm mã nhân viên đầu tiên có lổng lương lớn nhất.
# Câu 6. Cập nhật lại lương cơ bản theo mã nhân viên.
# Câu 7. Khởi tạo nhanh 6 nhân viên có dữ liệu sau (không dùng hàm input):
# (123, "Nguyen A", 4_500_000, 0.5, 50) #Chuyên viên
# (124, "Nguyen B", 5_600_000, 1.2, 10) #Nghiên cứu viên
# (125, "Nguyen C", 7_800_000, 1.5, 1.3) #Quản lý
# (126, "Nguyen D", 8_100_000, 0.8, 12) #Nghiên cứu viên
# (127, "Nguyen E", 9_500_000, 1.0, 1.6) #Quản lý
# (128, "Nguyen F", 6_500_000, 0.8, 30) #Chuyên viên

```python
from abc import ABC, abstractmethod

class abstractNV(ABC):
    def __init__(self,maNV,hoTen,luongCoBan,heSoTrachNhiem):
        self.maNV = maNV
        self.hoTen = hoTen
        self.luongCoBan = luongCoBan
        self.heSoTrachNhiem = heSoTrachNhiem
    
    @abstractmethod
    def xuat_thong_tin(self):
        pass
    
    @abstractmethod
    def tinh_luong(self):
        pass
    
class NV(abstractNV):
    def __init__(self,maNV,hoTen,luongCoBan,heSoTrachNhiem):
        super().__init__(maNV,hoTen,luongCoBan,heSoTrachNhiem)
        
    def xuat_thong_tin(self):
        print("Mã nhân viên:", self.maNV, ", ", end = '')
        print("Họ và tên:",self.hoTen, ", ", end = '')
        print("Lương cơ bản:",self.luongCoBan, ", ", end = '')
        print("Hệ số trách nhiệm:",self.heSoTrachNhiem, ", ", end = '')
    
    def tinh_luong(self):
        pass
    
class NVChuyenVien(NV):
    def __init__(self,maNV,hoTen,luongCoBan,heSoTrachNhiem,soGioOT):
        super().__init__(maNV,hoTen,luongCoBan,heSoTrachNhiem)
        self.soGioOT = soGioOT
    
    def xuat_thong_tin(self):
        super().xuat_thong_tin()
        print("Số giờ tăng ca:",self.soGioOT)
        print()
        
    def tinh_luong(self):
        return self.luongCoBan + self.luongCoBan * self.heSoTrachNhiem + self.soGioOT * 180000
    
class NVNghienCuuVien(NV):
    def __init__(self,maNV,hoTen,luongCoBan,heSoTrachNhiem,soCTPM):
        super().__init__(maNV,hoTen,luongCoBan,heSoTrachNhiem)
        self.soCTPM = soCTPM
        
    def xuat_thong_tin(self):
        super().xuat_thong_tin()
        print("Số công trình phát minh:",self.soCTPM)
        print()
        
    def tinh_luong(self):
        return self.luongCoBan + self.luongCoBan * (self.heSoTrachNhiem-0.2) + self.soCTPM * 5500000
    
class NVQuanLy(NV):
    def __init__(self,maNV,hoTen,luongCoBan,heSoTrachNhiem,heSoKiemNhiem):
        super().__init__(maNV,hoTen,luongCoBan,heSoTrachNhiem)
        self.heSoKiemNhiem = heSoKiemNhiem
        
    def xuat_thong_tin(self):
        super().xuat_thong_tin()
        print("Hệ số kiêm nhiệm:",self.heSoKiemNhiem)
        print()
        
    def tinh_luong(self):
        return self.luongCoBan * 0.7 + self.luongCoBan * self.heSoTrachNhiem + self.luongCoBan * self.heSoKiemNhiem
    
"""1. Xuất thông tin của các nhân viên 
(mã nhân viên, họ tên, lương cơ bản, hệ số trách nhiệm, tổng tiền lương), 
thông tin mỗi nhân viên xuất trên 1 dòng. """
nv1 = NVChuyenVien(123, "Nguyen Van A", 4500000, 0.5, 50)
nv2 = NVNghienCuuVien(124, "Nguyen Van B", 5600000, 1.2, 10)
nv3 = NVQuanLy(125, "Ha Thanh C", 7800000, 1.5, 1.3)
nv4 = NVChuyenVien(126, "Nguyen Tan D", 8100000, 0.5, 50)
nv5 = NVNghienCuuVien(127, "Banh Thi E", 9500000, 1.2, 10)
nv6 = NVQuanLy(128, "Nguyen F", 6500000, 1.5, 1.3)
dsNV = [nv1,nv2,nv3,nv4,nv5,nv6]
print("Danh sách các nhân viên:")
for i in dsNV:
    i.xuat_thong_tin()

"""Câu 2. Thực hiện việc tính lương cho từng nhân viên."""
print("DS lương cho từng nhân viên:")
for i in dsNV:
    i.xuat_thong_tin()
    print("Lương của nhân viên:", i.tinh_luong())

"""Câu 3. Tìm kiếm nhân viên theo mã nhân viên."""
searchID = int(input("Nhập mã nhân viên cần tìm: "))
flag=False
for i in dsNV:
    if i.maNV == searchID:
        flag = True
        i.xuat_thong_tin()
if flag == False:
    print("Không tìm thấy nhân viên")

"""Câu 4. Tính tổng số tiền phải trả cho tất cả nhân viên."""
def calSumSalary(dsNV):
    rs = 0
    for i in dsNV:
        rs += i.tinh_luong()
    return rs
print("Tổng lương cần phải trả cho toàn bộ nhân viên:",calSumSalary(dsNV))
"""Câu 5. Tìm mã nhân viên đầu tiên có tổng lương lớn nhất."""
def findMaxSalary(dsNV):
    maxS = 0
    for i in dsNV:
        if i.tinh_luong() > maxS:
            maxS = i.tinh_luong()
    return maxS

kq = findMaxSalary(dsNV)
print("Mã nhân viên đầu tiên có tổng lương lớn nhất:",end = '')
for i in dsNV:
    if i.tinh_luong() == kq:
        print(i.maNV)
        break
print()
"""Câu 6. Cập nhật lại lương cơ bản theo mã nhân viên."""
idOfNV = int(input("Nhập mã nhân viên cần cập nhật lương cơ bản"))
updateSalary = int(input("Nhập lương cơ bản bạn muốn cập nhật cho nhân viên đó:"))
for i in dsNV:
    if i.maNV == idOfNV:
        i.luongCoBan = updateSalary
        i.xuat_thong_tin()

"""Câu 7. Khởi tạo nhanh 6 nhân viên có dữ liệu sau (không dùng hàm input):
(123, "Nguyen A", 4_500_000, 0.5, 50) #Chuyên viên
(124, "Nguyen B", 5_600_000, 1.2, 10) #Nghiên cứu viên
(125, "Nguyen C", 7_800_000, 1.5, 1.3) #Quản lý
(126, "Nguyen D", 8_100_000, 0.8, 12) #Nghiên cứu viên
(127, "Nguyen E", 9_500_000, 1.0, 1.6) #Quản lý
(128, "Nguyen F", 6_500_000, 0.8, 30) #Chuyên viên"""
employees = [
    NVChuyenVien(123, "Nguyen A", 4_500_000, 0.5, 50), #Chuyên viên
    NVNghienCuuVien(124, "Nguyen B", 5_600_000, 1.2, 10), #Nghiên cứu viên
    NVQuanLy(125, "Nguyen C", 7_800_000, 1.5, 1.3), #Quản lý
    NVChuyenVien(126, "Nguyen D", 8_100_000, 0.8, 12), #Nghiên cứu viên
    NVNghienCuuVien(127, "Nguyen E", 9_500_000, 1.0, 1.6), #Quản lý
    NVQuanLy(128, "Nguyen F", 6_500_000, 0.8, 30) #Chuyên viên
]
employees[0].xuat_thong_tin()
```

# Bài 4.Xây dựng lớp đường tròn (DuongTron) được kế thừa từ lớp điểm (Diem) để biểu diễn một đường tròn trong mặt phẳng. Sau đó, viết chương trình minh họa:
# 1. Nhập đường tròn
# 2. Xuất thông tin đường tròn
# 3. Tính chu vi
# 4. Tính diện tích
# 5. Kiểm tra mối quan hệ giao, tiếp xúc, không giao của hai đường tròn

```python
import math

class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

class DuongTron(Diem):
    def __init__(self, x=0, y=0, r=0):
        Diem.__init__(self, x, y)
        self.r = r
        
    def chu_vi(self):
        return 2*math.pi*self.r
    
    def dien_tich(self):
        return math.pi*self.r**2
    
    def giao_nhau(self, dt):
        if math.sqrt((self.x-dt.x)**2 + (self.y-dt.y)**2) > self.r + dt.r:
            return "Hai đường tròn không giao nhau"
        elif math.sqrt((self.x-dt.x)**2 + (self.y-dt.y)**2) == self.r + dt.r:
            return "Hai đường tròn tiếp xúc ngoài"
        elif math.sqrt((self.x-dt.x)**2 + (self.y-dt.y)**2) == abs(self.r - dt.r):
            return "Hai đường tròn tiếp xúc trong"
        elif math.sqrt((self.x-dt.x)**2 + (self.y-dt.y)**2) < abs(self.r - dt.r):
            return "Hai đường tròn chứa nhau hoặc đè lên nhau"
        else:
            return "Hai đường tròn giao nhau"
        
"""1. Nhập đường tròn"""
dt1 = DuongTron(0, 0, 5)
dt2 = DuongTron(10, 0, 3)
dt3 = DuongTron(20, 0, 4)

"""2. Xuất thông tin đường tròn"""
print("Thông tin đường tròn 1:")
print("Tâm:", dt1.x, dt1.y)
print("Bán kính:", dt1.r)
print("Chu vi:", dt1.chu_vi())
print("Diện tích:", dt1.dien_tich())
print()

print("Thông tin đường tròn 2:")
print("Tâm:", dt2.x, dt2.y)
print("Bán kính:", dt2.r)
print("Chu vi:", dt2.chu_vi())
print("Diện tích:", dt2.dien_tich())
print()

print("Thông tin đường tròn 3:")
print("Tâm:", dt3.x, dt3.y)
print("Bán kính:", dt3.r)
print("Chu vi:", dt3.chu_vi())
print("Diện tích:", dt3.dien_tich())
print()

"""3. Tính chu vi"""

print("Thông tin đường tròn 1:")
print("Tâm:", dt1.x, dt1.y)
print("Bán kính:", dt1.r)
print("Chu vi:", dt1.chu_vi())
print()

print("Thông tin đường tròn 2:")
print("Tâm:", dt2.x, dt2.y)
print("Bán kính:", dt2.r)
print("Chu vi:", dt2.chu_vi())
print()

print("Thông tin đường tròn 3:")
print("Tâm:", dt3.x, dt3.y)
print("Bán kính:", dt3.r)
print("Chu vi:", dt3.chu_vi())
print()

"""4. Tính diện tích"""
print("Thông tin đường tròn 1:")
print("Tâm:", dt1.x, dt1.y)
print("Bán kính:", dt1.r)
print("Diện tích:", dt1.dien_tich())
print()

print("Thông tin đường tròn 2:")
print("Tâm:", dt2.x, dt2.y)
print("Bán kính:", dt2.r)
print("Diện tích:", dt2.dien_tich())
print()

print("Thông tin đường tròn 3:")
print("Tâm:", dt3.x, dt3.y)
print("Bán kính:", dt3.r)
print("Diện tích:", dt3.dien_tich())
print()

"""5. Kiểm tra mối quan hệ giao, tiếp xúc, không giao của hai đường tròn"""
print(dt1.giao_nhau(dt2))
print(dt1.giao_nhau(dt3))
print(dt2.giao_nhau(dt3))
```

# Bài 5. Xây dựng lớp hỗn số (HonSo) được kế thừa từ lớp phân số (PhanSo) để biểu diễn một hỗn số gồm hai thành phần:
# - Thành phần dữ liệu: phần nguyên, tử số, mẫu số.
# - Thành phần xử lý: xuất-nhập giá trị cho hỗn số, chuyển đổi hỗn số sang phân số và ngược lại, cộng, trừ, nhân, chia hai hỗn số, so sánh hai hỗn số. Chú ý: vận dụng overload (+, -, *, /, ==, !=, >, >=, <, <=)
# Sau đó, thiết kế chương trình cho phép:
# 1. Khởi tạo hỗn số
# 2. Xuất hỗn số
# 3. Chuyển đổi phân số sang hỗn số
# 4. Chuyển đổi hỗn số sang phân số
# 5. Thực hiện các phép toán cộng, trừ, nhân, chia
# 6. Thực hiện phép so sánh.
# 7. Tự xây dựng thêm các thao tác thích hợp minh họa cho bài toán.

```python
class PhanSo:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau

    def __str__(self):
        return str(self.tu) + '/' + str(self.mau)

    def doi_dau(self):
        self.tu = -self.tu

    def rut_gon(self):
        ucln = self.UCLN(self.tu, self.mau)
        self.tu //= ucln
        self.mau //= ucln

    def __add__(self, other):
        tu_moi = self.tu*other.mau + other.tu*self.mau
        mau_moi = self.mau * other.mau
        return PhanSo(tu_moi, mau_moi)

    def __sub__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        mau_moi = self.mau * other.mau
        return PhanSo(tu_moi, mau_moi)

    def __mul__(self, other):
        tu_moi = self.tu * other.tu
        mau_moi = self.mau * other.mau
        return PhanSo(tu_moi, mau_moi)

    def __truediv__(self, other):
        tu_moi = self.tu * other.mau
        mau_moi = self.mau * other.tu
        return PhanSo(tu_moi, mau_moi)

    def __eq__(self, other):
        self.rut_gon()
        other.rut_gon()
        return self.tu == other.tu and self.mau == other.mau

    def __ne__(self, other):
        return not self == other

    def __gt__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        return tu_moi > 0

    def __ge__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        return tu_moi >= 0

    def __lt__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        return tu_moi < 0

    def __le__(self, other):
        tu_moi = self.tu*other.mau - other.tu*self.mau
        return tu_moi <= 0

    def UCLN(self, a, b):
        if b == 0:
            return a
        return self.UCLN(b, a%b)
    
    def to_decimal(self):
        return self.tu / self.mau
    
    def inverse(self):
        return PhanSo(self.mau, self.tu)
    
    def power(self, n):
        return PhanSo(self.tu ** n, self.mau ** n)
    
    def chuyen_doi_hon_so(self):
        phan_nguyen = self.tu // self.mau
        phan_du = self.tu % self.mau
        return HonSo(phan_nguyen, phan_du, self.mau)
    
class HonSo(PhanSo):
    def __init__(self,phan_nguyen,tu,mau):
        super().__init__(tu,mau)
        self.phan_nguyen = phan_nguyen
        
    def __str__(self):
        return f"{self.phan_nguyen} {super().__str__()}"

    def xuat(self):
        print(self)

    def chuyen_doi_phan_so(self):
        tu_moi = self.phan_nguyen * self.mau + self.tu
        return PhanSo(tu_moi, self.mau)
    
    def __add__(self, other):
        tu_so = self.phan_nguyen * self.mau + self.tu
        mau_so = self.mau
        tu_so += other.phan_nguyen * other.mau + other.tu
        mau_so = max(mau_so, other.mau)
        return HonSo(tu_so, 0, mau_so)

    def __sub__(self, other):
        tu_so = self.phan_nguyen * self.mau + self.tu
        mau_so = self.mau
        tu_so -= other.phan_nguyen * other.mau + other.tu
        mau_so = max(mau_so, other.mau)
        return HonSo(tu_so, 0, mau_so)

    def __mul__(self, other):
        tu_so = (self.phan_nguyen * self.mau + self.tu) * (other.phan_nguyen * other.mau + other.tu)
        mau_so = self.mau * other.mau
        return HonSo(tu_so, 0, mau_so)

    def __truediv__(self, other):
        tu_so = (self.phan_nguyen * self.mau + self.tu) * other.mau
        mau_so = self.mau * (other.phan_nguyen * other.mau + other.tu)
        return HonSo(tu_so, 0, mau_so)
    
    def __eq__(self, other):
        ps = self.chuyen_doi_phan_so()
        ps1 = other.chuyen_doi_phan_so()
        ps.rut_gon()
        ps1.rut_gon()
        return ps.tu == other.tu and ps1.mau == other.mau

    def __ne__(self, other):
        return not self == other

    def __gt__(self, other):
        ps = self.chuyen_doi_phan_so()
        ps1 = other.chuyen_doi_phan_so()
        tu_moi = ps.tu*ps1.mau - ps1.tu*ps.mau
        return tu_moi > 0

    def __ge__(self, other):
        ps = self.chuyen_doi_phan_so()
        ps1 = other.chuyen_doi_phan_so()
        tu_moi = ps.tu*ps1.mau - ps1.tu*ps.mau
        return tu_moi >= 0

    def __lt__(self, other):
        ps = self.chuyen_doi_phan_so()
        ps1 = other.chuyen_doi_phan_so()
        tu_moi = ps.tu*ps1.mau - ps1.tu*ps.mau
        return tu_moi < 0

    def __le__(self, other):
        ps = self.chuyen_doi_phan_so()
        ps1 = other.chuyen_doi_phan_so()
        tu_moi = ps.tu*ps1.mau - ps1.tu*ps.mau
        return tu_moi <= 0
    
    def to_decimal(self):
        ps = self.chuyen_doi_phan_so()
        return ps.tu / ps.mau
    
    def power(self, n):
        return HonSo(self.phan_nguyen ** n,self.tu ** n, self.mau ** n)

"""1. Khởi tạo hỗn số """
a = HonSo(1,2,3)
b = HonSo(3,1,3)

"""2. Xuất hỗn số"""
print("Các hỗn số:")
a.xuat()
b.xuat()
"""3. Chuyển đổi phân số sang hỗn số"""
print("Phân số sau khi chuyển đổi:")
ps1 = PhanSo(7,2)
c = ps1.chuyen_doi_hon_so()
c.xuat()
"""4. Chuyển đổi hỗn số sang phân số"""
print("Hỗn số sau khi chuyển đổi sang phân số:")
d = HonSo(2,5,6)
ps2 = d.chuyen_doi_phan_so()
print(ps2)
"""5. Thực hiện các phép toán cộng, trừ, nhân, chia"""
d = a + b
print("a + b = ", d)

e=a-b
print("a - b = ",e)

f=a*b
print("a * b = ",f)

g=a/b
print("a / b = ",g)
"""6. Thực hiện phép so sánh."""
print("a > b = ",a>b)
"""7. Tự xây dựng thêm các thao tác thích hợp minh họa cho bài toán."""
hs = HonSo(4,5,6)
print("Phân số:", hs)
print("Số thập phân:", hs.to_decimal())
print("Lũy thừa bậc 3 của phân số:", hs.power(3))
```

```python
class SinhVien:
    def __init__(self, ma_sv, ho_ten, diem_tb, tinh_trang):
        self.ma_sv = ma_sv
        self.ho_ten = ho_ten
        self.diem_tb = diem_tb
        self.tinh_trang = tinh_trang

    def display(self):
        print("Mã sinh viên: ", self.ma_sv)
        print("Họ và tên: ", self.ho_ten)
        print("Điểm trung bình: ", self.diem_tb)
        print("Tình trạng: ", self.tinh_trang)


class LopHoc:
    def __init__(self,tenLop):
        self.tenLop = tenLop
        self.danh_sach_sv = []

    def them_sinh_vien(self, sv):
        if len(self.danh_sach_sv) < 50:
            self.danh_sach_sv.append(sv)
            return True
        else:
            return False

    def nhap_sinh_vien(self):
        ma_sv = input("Nhập mã sinh viên: ")
        ho_ten = input("Nhập họ và tên: ")
        diem_tb = float(input("Nhập điểm trung bình: "))
        tinh_trang = input("Nhập tình trạng (đang học hoặc thôi học): ")
        sv = SinhVien(ma_sv, ho_ten, diem_tb, tinh_trang)
        return sv

    def hien_thi_danh_sach(self):
        for sv in self.danh_sach_sv:
            sv.display()

    def tao_du_lieu_ngau_nhien(self, so_luong):
        import random
        ma_sv_list = ["SV{:04d}".format(i) for i in range(so_luong)]
        ho_ten_list = ["Nguyen Van A", "Tran Thi B", "Le Van C", "Pham Thi D", "Hoang Van E"]
        diem_tb_list = [round(random.uniform(0, 10), 2) for _ in range(so_luong)]
        tinh_trang_list = ["đang học", "thôi học"]
        for i in range(so_luong):
            ma_sv = random.choice(ma_sv_list)
            ho_ten = random.choice(ho_ten_list)
            diem_tb = random.choice(diem_tb_list)
            tinh_trang = random.choice(tinh_trang_list)
            sv = SinhVien(ma_sv, ho_ten, diem_tb, tinh_trang)
            self.them_sinh_vien(sv)

"""
1. Nhập, lưu trữ nhiều sinh viên. Biết mỗi lớp có tối thiểu 30 sinh viên và
tối đa 50 sinh viên.
2. Cho phép tạo ngẫu nhiên dữ liệu nhiều sinh viên dùng để kiểm thử.
"""
lopA = LopHoc("Lớp A")
lopB = LopHoc("Lớp B")
lopC = LopHoc("Lớp C")

a = int(input("Nhập số sinh viên cần tạo: "))
if a>=30 and a<=50:
    lopA.tao_du_lieu_ngau_nhien(a)
else:
    print("Bạn nhập sai số sinh viên yêu cầu của 1 lớp.")
    
b = int(input("Nhập số sinh viên cần tạo: "))
if b>=30 and b<=50:
    lopB.tao_du_lieu_ngau_nhien(b)
else:
    print("Bạn nhập sai số sinh viên yêu cầu của 1 lớp.")

c = int(input("Nhập số sinh viên cần tạo: "))
if c>=30 and c<=50:
    lopC.tao_du_lieu_ngau_nhien(c)
else:
    print("Bạn nhập sai số sinh viên yêu cầu của 1 lớp.")    
"""3. Xuất các sinh viên."""
print("Danh sách sinh viên lớp A:")
lopA.hien_thi_danh_sach()
print()
print("Danh sách sinh viên lớp B")
lopB.hien_thi_danh_sach()
print()
print("Danh sách sinh viên lớp C")
lopC.hien_thi_danh_sach()
print()
"""4. Tìm sinh viên theo mã sinh viên."""
ds = []
for i in lopA.danh_sach_sv:
    ds.append(i)
for i in lopB.danh_sach_sv:
    ds.append(i)
for i in lopC.danh_sach_sv:
    ds.append(i)

searchID = input("Nhập mã SV cần tìm: ")
flag = False
for i in ds:
    if i.ma_sv == searchID:
        flag = True
        i.display()
        print()
if(flag == False):
    print("Không tìm thấy sinh viên")
"""5. Xuất các sinh viên đang thôi học."""
print ("Danh sách SV đang thôi học:")
for i in ds:
    if i.tinh_trang == "thôi học":
        i.display()
        print()
"""6. Tìm sinh viên có điểm TB cao nhất nhất trong lớp"""
def findMaxGPA(lop):
    rs = 0
    for i in lop.danh_sach_sv:
        if i.diem_tb > rs:
            rs = i.diem_tb
    return rs

maxA = findMaxGPA(lopA)
maxB = findMaxGPA(lopB)
maxC = findMaxGPA(lopC)

print("SV cao điểm nhất lớp A:")
for i in lopA.danh_sach_sv:
    if i.diem_tb == maxA:
        i.display()
        print()

print("SV cao điểm nhất lớp B:")
for i in lopB.danh_sach_sv:
    if i.diem_tb == maxB:
        i.display()
        print()
        
print("SV cao điểm nhất lớp C:")
for i in lopC.danh_sach_sv:
    if i.diem_tb == maxC:
        i.display()
        print()

"""7. Sắp xếp các lớp theo số lượng sinh viên theo chiều giảm dần."""
dsL = [lopA,lopB,lopC]
dsLsorted = sorted(dsL, key=lambda x: len(x.danh_sach_sv), reverse=True)
for i in dsLsorted:
    print(i.tenLop)
```
